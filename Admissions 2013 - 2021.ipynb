## 2013-2019 Admissions

```python
import pandas as pd
import datetime
pd.set_option('display.max_columns', 600)
```

```python
url1 = "https://raw.githubusercontent.com/itsyanlee/DS4A_data/admissions/2013-2019_Attendance_Results_-_School.csv"
admis_2013_2019 = pd.read_csv(url1, index_col = 0)

url2 = "https://raw.githubusercontent.com/itsyanlee/DS4A_data/admissions/2018-2021_Daily_Attendance_by_School.csv"
admis_2018_2021 = pd.read_csv(url2, index_col = 0)
```

```python
admis1 = admis_2013_2019.copy()
admis2 = admis_2018_2021.copy()
admis2 = admis2[(admis2['Date'].str.contains('2020')) | (admis2['Date'].str.contains('2021'))]
admis1
```

```python
a = admis2['SchoolYear'].astype(str)
clean_year = []

for x in a:
    list1 = x[:4]
    list2 = x[6:8]
    clean_year.append(list1+'-'+list2)
    
admis2['SchoolYear'] = clean_year
```

```python
admis2.rename(columns = {'SchoolYear': 'Year'}, inplace = True)
admis2.index.name = 'DBN'
```

```python
admis2_total_days = admis2.groupby(['DBN', 'Year'])[['Date','Year','Enrolled','Present','Absent','Released']].sum()

admis2_total_days['% Attendance'] = ((admis2_total_days.Present / admis2_total_days.Enrolled)*100).round(1)
```

```python
admis2_total_days.drop(columns = {'Enrolled','Present','Absent','Released' }, inplace = True)
```

```python
#only grades 5-8
admis1 = admis1.loc[(admis1.Grade == '5')| (admis1.Grade == '6') | (admis1.Grade == '7') | (admis1.Grade == '8')]
admis1.drop(columns = {'Demographic Category','Demographic Variable','# Total Days','# Days Absent','# Days Present', '# Contributing 20+ Total Days', '# Chronically Absent', '% Chronically Absent', "School Name"}, inplace = True)
```

```python
admis2_total_days = admis2_total_days.reset_index()
admis1 = admis1.reset_index()
```

```python
admissions = pd.concat([admis1,admis2_total_days] )
admissions.set_index(['DBN'], inplace = True)
```

```python
admissions.to_csv('admissions_2013-2022')
```
